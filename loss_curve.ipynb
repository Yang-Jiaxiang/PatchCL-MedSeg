```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'supervised pre training_loss_0510.csv'  # Replace with your actual file path
data = pd.read_csv(file_path)

# Assuming the 'loss' column is directly associated with epochs
plt.figure(figsize=(6, 4))  # Set figure size
plt.plot(data['loss'], color='blue')  # Plot the 'loss' column
plt.xlabel('Epoch')  # Label x-axis as 'Epoch'
plt.ylabel('Loss')  # Label y-axis as 'Loss'
plt.title('Loss Curve')  # Give the plot a title
plt.grid(True)  # Show grid

# Show the plot
plt.show()
```
